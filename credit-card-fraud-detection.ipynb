Importing Libraries

```python
import pandas as pd

from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data.drop(columns=["Unnamed: 0", "trans_num", "street"], inplace= True)
data
```

```python
df = data.head(n = 20000) # selecting first 20,000 rows
df.is_fraud.value_counts() # counting

# 0 - non-fraud
# 1 - fraud
```

One-Hot Encoding

- For example, if data has a column color with values ['red', 'blue', 'green'], pd.get_dummies(data=data) would replace this column with three new columns: color_red, color_blue, and color_green, where each column contains binary values indicating the presence of each color.

```python
df_processed = pd.get_dummies(data=df) 
df_processed
```

```python
x_train = df_processed.drop(columns='is_fraud', axis=1)
y_train = df_processed['is_fraud']
```
